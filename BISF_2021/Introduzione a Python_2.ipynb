### Matrici e vettori

```python
# per creare e operare su matrici e vettori occorre importare Numpy
import numpy as np
```

Una matrice è una tabella rettangolare di numeri.
Di default il comando di creazione di una matrice in Numpy assume che i dati siano organizzati per colonne
Le dimensioni di una matrice sono date numero righe (m) e numero  colonne (n) mxn

#### Comandi di Numpy

```python
a=np.zeros(10)                    # array with 10 zeros
b=np.zeros((3,2),dtype=float)     # 3 by 2 with zeros
c=np.ones((4,3),float)            # 4 by 3 with all ones
d=np.array(range(10),float)       # 0,1, 2,3 .. up to 9
e1=np.identity(4)                 # identity 4 by 4 matrix
e2=np.eye(4)                      # same as above
e3=np.eye(4,k=1)                  # 1 start from k
f=np.arange(1,20,3,float)         # from 1 to 19 interval 3
g=np.array([[2,2,2],[3,3,3]])     # 2 by 3
h=np.zeros_like(g)                # all zeros
i=np.ones_like(g)                 # all ones

# flatten
x=np.array([[1,2],[5,6],[7,9]])     # a 3 by 2 array
print(x)
y=x.flatten()
print(y)
x2=np.reshape(y,[2,3])
print(x2)

# dotproduct
import numpy as np
a=np.array([[1,2,3],[4,5,6]],float)   # 2 by 3
b=np.array([[1,2],[3,3],[4,5]],float) # 3 by 2
np.dot(a,b)                           # 2 by 2
print(np.dot(a,b))
```

### Liste

Lists are used to store multiple items in a single variable.

Lists are one of 4 built-in data types in Python used to store collections of data, the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.

Lists are created using square brackets:
thislist = ["apple", "banana", "cherry"]
print(thislist)

List items are ordered, changeable, and allow duplicate values.

List items are indexed, the first item has index [0], the second item has index [1] etc.
To determine how many items a list has, use the len() function


fonte: https://www.w3schools.com/python/python_lists.asp

```python
a = list([2, True, "hello"])
a
```

```python
# ma anche, più semplicemente

b = [2, True, "hello"]
b
```

Creazione di una lista con un ciclo for

```python
c = []

for i in range(1,20):
        c.append(i)

print(c)
```

### Data Frame

Il dataframe è utilizzato per archiviare dati in forma tabulare
Può essere visto come un insieme di liste 
Il data frame è la struttura di dati create dalla libreria Pandas ("Panles Data Series")
Two-dimensional, size-mutable, potentially heterogeneous tabular data.
Data structure also contains labeled axes (rows and columns). Arithmetic operations align on both row and column labels. Can be thought of as a dict-like container for Series objects. The primary pandas data structure.

fonte: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html

#### Operazioni su data frame

Si può lavorare sui DF utilizzando indici o nomi di righe/colonne

```python
price1= df.iloc[0,0]
price1
```

The main distinction between loc and iloc is:
loc is label-based, which means that you have to specify rows and columns based on their row and column labels.
iloc is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

fonte https://towardsdatascience.com/how-to-use-loc-and-iloc-for-selecting-data-in-pandas-bd09cb4c3d79

```python
price1 = df.iloc[0:2,0]
price1
```

```python
price1 = df.iloc[0:2]['Price']
price1
```
